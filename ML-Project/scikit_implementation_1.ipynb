Dataset: Breast Cancer





```
from sklearn.datasets import load_breast_cancer # breast cancerin scikit learn data set
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler  # It is important in neural networks to scale the date
from sklearn.model_selection import train_test_split  # The standard - train/test to prevent overfitting and choose hyperparameters
from sklearn.metrics import accuracy_score # 
import numpy as np
import numpy.random as r # We will randomly initialize our weights
import matplotlib.pyplot as plt 

breast_cancer=load_breast_cancer()
X = breast_cancer.data
print("The shape of the breast cancer dataset:") 
print(breast_cancer.data.shape)
y = breast_cancer.target

X_scale = StandardScaler()
X = X_scale.fit_transform(breast_cancer.data)

#Split the data into training and test set.  60% training and %40 test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

sklearn logistic regression on breast cancer dataset (i)

```
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(penalty='none',solver='saga',tol=0,max_iter=5000).fit(X_train, y_train)
clf.predict(X_test)
clf.predict_proba(X_test)
print(clf.score(X_test, y_test))
```

sklearn logistic regression + LASSO regularization (extension) on breast cancer dataset (ii)

```
clf = LogisticRegression(penalty='l1',solver='saga',tol=0,max_iter=5000).fit(X_train, y_train)
clf.predict(X_test)
clf.predict_proba(X_test)
print(clf.score(X_test, y_test))
```

Dataset: Wine (slightly modified for binary classification, see below)

```
from sklearn.datasets import load_wine # wine in scikit learn data set
from sklearn import preprocessing # It is important in neural networks to scale the date
from sklearn.model_selection import train_test_split  # The standard - train/test to prevent overfitting and choose hyperparameters
from sklearn.metrics import accuracy_score # 
import numpy as np
import numpy.random as r # We will randomly initialize our weights
import matplotlib.pyplot as plt 

wine=load_wine()
X = wine.data
print("The shape of the wine dataset:") 
print(wine.data.shape)
y = wine.target

X = np.vstack([X[y==1],X[y==2]])
y = np.hstack([y[y==1],y[y==2]])
print("The new shape of the wine dataset:") 
print(X.shape)

#Split the data into training and test set.  60% training and %40 test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

sklearn logistic regression on wine dataset (iii)

```
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(penalty='none',solver='saga',tol=0,max_iter=5000).fit(X_train, y_train)
clf.predict(X_test)
clf.predict_proba(X_test)
print(clf.score(X_test, y_test))
```

sklearn logistic regression + LASSO regularization (extension) on wine dataset (iv)

```
clf = LogisticRegression(penalty='l1',solver='saga',tol=0,max_iter=5000).fit(X_train, y_train)
clf.predict(X_test)
clf.predict_proba(X_test)
print(clf.score(X_test, y_test))
```
